```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()

print(boston.DESCR)
```

```python
# should always do a quick visualization of the data 
# create histogram of prices.  prices is the current 'target' or 'label' of our dataset
# creating 50 bins because there are 506 number of instances

plt.hist(boston.target, bins=50)

plt.xlabel('Prices in $1000s')
plt.ylabel('Number of Houses')
```

```python
# scatter plot one feature versus a target
# look at rooms of dwelling. 6 down in the attribute information list so it will be index 5

# x = everything in the 6th column, y = price (or the target)
plt.scatter(boston.data[:,5], boston.target)

plt.ylabel('Price in $1000s')
plt.xlabel('Number of Rooms')

# will see a positive correlation between number of rooms and price of house 
```

```python
#use pandas to transform Boston dataset into a DataFrame 
# reset the data as a pandas dataframe

boston_df = DataFrame(boston.data)
boston_df
```

```python
boston_df.columns=boston.feature_names
boston_df

# now you've created a dataframe with all the attributes or features
```

```python
# add target of the boston dataset, the price

boston_df['Price'] = boston.target
boston_df
```

```python
# Seaborn lmplot feature can be used to do a linear regression automatically

sns.lmplot('RM','Price',data=boston_df)
```
