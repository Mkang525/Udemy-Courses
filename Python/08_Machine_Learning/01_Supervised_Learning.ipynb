```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()

print(boston.DESCR)
```

```python
# should always do a quick visualization of the data 
# create histogram of prices.  prices is the current 'target' or 'label' of our dataset
# creating 50 bins because there are 506 number of instances

plt.hist(boston.target, bins=50)

plt.xlabel('Prices in $1000s')
plt.ylabel('Number of Houses')
```

```python
# scatter plot one feature versus a target
# look at rooms of dwelling. 6 down in the attribute information list so it will be index 5

# x = everything in the 6th column, y = price (or the target)
plt.scatter(boston.data[:,5], boston.target)

plt.ylabel('Price in $1000s')
plt.xlabel('Number of Rooms')

# will see a positive correlation between number of rooms and price of house 
```

```python
#use pandas to transform Boston dataset into a DataFrame 
# reset the data as a pandas dataframe

boston_df = DataFrame(boston.data)
boston_df
```

```python
boston_df.columns=boston.feature_names
boston_df

# now you've created a dataframe with all the attributes or features
```

```python
# add target of the boston dataset, the price

boston_df['Price'] = boston.target
boston_df
```

```python
# Seaborn lmplot feature can be used to do a linear regression automatically

sns.lmplot('RM','Price',data=boston_df)
```

```python
# using Numpy for a Univariate Linear Regression (Univariate = Single Variable)
# manually create what we just did above using numpy

#this way only gives you values (506)
X = boston_df.RM
print(X)

X.shape
```

```python
# numpy needs to know how many values and how many attributes.  
# so here we have 1 attribute, the room size.  And 506 values for that attribute
# vstack will stack arrays in sequence vertically (row wise).

X = np.vstack(boston_df.RM)
# X = np.vstack([boston_df.RM,np.ones(len(boston_df.RM))]).T
X.shape
```

```python
Y = boston_df.Price
Y
```

```python
# [X 1]--> needs to look like this
X
```

```python
#using a list comprension.  another way to do a for loop that builds a list

# for each value in X, set it up as [value,1] and then make an array(np.array) of all those values
X = np.array( [ [value,1] for value in X] )
X
```

```python
# need to make array a float not an object

X = X.astype(np.float64)
```

```python
# now we can get our best fit values
# get m and b values for best fit line
# numpy method in the linear algebra library (linalg) and then we're using the least squares method (lstq)
# we only want the 1st index

m,b = np.linalg.lstsq(X,Y, rcond=None)[0]
m,b
```

```python
#create a scatter plot
plt.plot(boston_df.RM, boston_df.Price, 'o')

# plot best fit line we just made using numpy
x = boston_df.RM

plt.plot(x, m*x + b, 'r',label='Best Fit Line')
```

```python
# now we can find the error in the fitted line.  -- > the difference between the line and the true value. (D*2), squares it and returns the sum of all of it
# similar to the standard deviation 

result = np.linalg.lstsq(X,Y, rcond=None)

error_total = result[1]

rmse = np.sqrt(error_total/len(X))

print('The root mean square error was %.2f.' %rmse)
print(' We can reasonably expect a house price to be within $13,200 of our line fit.')
```

# Step 6: Using scikit learn to implement a multivariate regression

<p>Now, we'll keep moving along with using scikit learn to do a multi variable regression. This will be a similar apporach to the above example, but sci kit learn will be able to take into account more than just a single data variable effecting the target!

We'll start by importing the linear regression library from the sklearn module.</p>

<ul>
    <li>The sklearn.linear_model.LinearRegression class is an estimator. </li>
    <li>Estimators predict a value based on the observed data. </li>
    <li>In scikit-learn, all estimators implement the fit() and predict() methods. </li>
    <li>The fit method is used to learn the parameters of a model, and the predict method is used to predict the value of a response variable for an explanatory variable using the learned parameters. </li>
    <li>It is easy to experiment with different models using scikit-learn because all estimators implement the fit and predict methods.</li>
    </ul>
    
<a href="https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html">scikit learn library</a>

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

<h3> Functions we'll be using</h3>

<ul>
    <li>lreg.fit() which fits a linear model</li>
    <li>lreg.predict() which is used to predict Y using the linear model with estimated coefficients</li>
    <li>lreg.score() which returns the coefficient of determination (R^2)</li>
    <li>R^2 is a measure of how well observed outcomes are replicated by the model. </li>
</ul>



```python
#Create a LinearRegression Object
lreg = LinearRegression()
```

```python
# in order to drop a column, have to pass a 1 index

X_multi = boston_df.drop('Price',1)

Y_target = boston_df.Price
```

```python
# pass the X and Y using the linear regression object

lreg.fit(X_multi,Y_target)
```

```python
# check the intercept and the number of coefficients

print('The estimated intercept coefficient is %.2f ' %lreg.intercept_)

print('The number of coefficients used %d ' %len(lreg.coef_))
```

```python
# Set up a DataFrame showing all the Features and their estimated coefficients obtained from the linear regression

#create a new datafrae based on the columns of the original Boston df
coeff_df = DataFrame(boston_df.columns)

#Rename the column 'Features'
coeff_df.columns=['Features']

# Created a new column and passed all the coefficents in the new column
coeff_df['Coefficient Estimate'] = Series(lreg.coef_)
coeff_df
```

Just like we initially plotted out, it seems the highest correlation between a feature and a house price was the number of rooms. (RM = ROOMS, has the highest coeff @ 3.81
