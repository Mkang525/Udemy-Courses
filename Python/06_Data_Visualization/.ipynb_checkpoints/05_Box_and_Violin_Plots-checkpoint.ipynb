```python
import numpy as np
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import webbrowser
url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg'

webbrowser.open(url)
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2])
```

```python
#if you want to include outliers
#np.inf = infinity

sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# violin plots combines simple format of box plot with the info of a KDE plot
#dont worry about understanding below , just an example

#Normal Dist
data1 = stats.norm(0,5).rvs(100)

#Two gamma dist. concatenate them together 
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

#Box plot both data1 and data2
sns.boxplot(data=[data1,data2], whis = np.inf)
```

```python
# violin plots allow you to see the bigger picture than just a box plot.  It combines a KDE plot with a box plot

sns.violinplot(data=[data1,data2])
```

```python
# can also change the bandwidth 

sns.violinplot(data2,bw=0.01)
```

```python
# can include the individual plots

sns.violinplot(data1, inner='stick')
```
