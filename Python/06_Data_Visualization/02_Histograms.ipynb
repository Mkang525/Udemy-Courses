```python
dataset1 = randn(100)
```

```python
#default number of bins is 10 

plt.hist(dataset1)
```

```python
dataset2 = randn(80)

plt.hist(dataset2, color='indianred')
```

```python
#combining the 2 histograms together
# because they have 2 different data points, have to normalize the plots (density=True) 
# alpha means transparency

plt.hist(dataset1,density=True, color='indianred',alpha=0.5, bins=20)
plt.hist(dataset2,density=True, alpha=0.5, bins=20)
```

```python
#Using Seaborn
data1 = randn(1000)
data2 = randn(1000)
```

```python
#can be represented as a joint distribution using joint plots.

sns.jointplot(data1,data2)
```

```python
sns.jointplot(data1,data2, kind='hex')
```
