# Numpy Where

```python
A = np.array([1,2,3,4])

B = np.array([100,200,300,400])
```

```python
#creating a boolean array
condition = np.array([True, True,False,False])
```

```python
#in this array show me A_val if the condition is met, otherwise show me the B_val
#defines A_val, B_val, cond as the value in A, B, condition

answer = [(A_val if cond else B_val) for A_val, B_val, cond in zip(A,B,condition)]
```

```python
#only lists 1 and 2 from 'A' because they are true in condition. 
#showing 300 and 400 from 'B' because it is false in condition
answer
```

```python
#numpy where is the short version of above

answer2 = np.where(condition,A,B)
answer2
```
